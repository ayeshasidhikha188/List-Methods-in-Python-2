##  6. Sort()
* The sort function can be used to sort the list in both ascending and descending order.
* It can be used to sort lists of integers, floating point numbers, strings,
* Syntax: list_name.sort(key=, reverse=)
##### Parameters:
* By default, sort() doesnt require any extra parameters and sorts the list in ascending order. However, it has two optional parameters
* key:  function that serves as a key for the sort comparison
* reverse: If true, the list is sorted in descending order.
##### Return value:
* The sort() does not return anything but alters the original list according to the passed parameter.


#### Sorting List in Ascending Order
* By default, the sort()  sorts a list in ascending order if we do not provide it with any parameters.

```python
# Sort() 
x,y=[1,-5,10,14,100,89],['a','ashu','b','z','ayesha']
z=x+y #unpacking or concatenation(adding 'x' and 'y') #sequence is first x and then y 
print(z)
```

```python
x.sort()
print(x) #  By default it takes ascending order
```

```python
y.sort()
print(y)
```

```python
a=['ayesha','arun','a.prasanna','abhi','aaa','AAA']
a.sort()
print(a) # depending on ascii values
```

#### Sorting List in Descending Order
* To sort a list in descending order, set the reverse parameter to True of the sort() function 

```python
# to convert int descending order reverse=true 
#reverse=false is for ascending order
aa=[10,-80,960,1000,89]
aa.sort(reverse=True)
print(aa) # output is descending order (higher to lower)
```

## 10.  copy()
* Copy() methods return a shallow copy of a list, which returns a new list without modifying the original lists.
* Sometimes, there is a need to reuse any object, hence copy methods are always of great utility. 
* Since the list is widely used, its copy is also necessary.
* Syntax: list.copy()
##### Parameters: 
* The copy() method doesnt take any parameters
##### Returns:
* Returns a shallow copy of a list.
* A shallow copy means any modification in the new list wont be reflected in the original list. 


```python
x=[11,22,33]
print(x,hex(id(x)))
```

```python
y=x
print(y,hex(id(y))) # x &y have same memory location
```

```python
print(hex(id(x[0])),hex(id(y[0]))) #same memory location
```

```python
z=x.copy()
print(z,hex(id(z)))
```

### Deep copy()
* A deep copy creates a new compound object before inserting copies of the items found in the original into it in a recursive manner. 
* In the case of deep copy, a copy of the object is copied into another object. 
* It means that any changes made to a copy of the object do not reflect in the original object. 
* It is used to copy  nested list


```python
a=[10,[45,46,47]]
print(a,a[1],id(a[1]))
```

```python
b=a.copy()
print(b,b[1],id(b[1])) #same memory location 
```

```python
# to handle nested list deep copy is used
import copy
from copy import deepcopy
```

```python
a=[10,[45,46,47]]
b=deepcopy(a)
print(a,a[1],id(a[1])) # different memory location 
print(b,b[1],id(b[1]))
```
